```python
#Import thư viện pandas
import pandas as pd
```

```python
air_quality = pd.read_csv('historical_air_quality_2021_en (1).csv')
print(air_quality.columns)
```

```python
#Kiểm tra dạng dữ liệu của các cột
air_quality.info()
```

```python
#Chuyển dữ liệu về dạng date time
air_quality['Data Time S'] = pd.to_datetime(air_quality['Data Time S'],utc=True)
```

```python
air_quality.info()
```

```python
#Tách cột date time s thành các cột time, date, year, month, day
def split_date(air_quality,col):
    time = []
    date = []
    year = []
    month = []
    day = []
    i = 0
    
    for i in range(air_quality.shape[0]):
        t = air_quality[col][i].time()
        d = air_quality[col][i].date()
        y = air_quality[col][i].date().year
        m = air_quality[col][i].date().month
        da = air_quality[col][i].date().day
        time.append(t)
        date.append(d)
        year.append(y)
        month.append(m)
        day.append(da)
        i = i+1
    
    air_quality['date'] = date
    air_quality['time'] = time
    air_quality['year'] = year
    air_quality['month'] = month
    air_quality['day'] = day
    
split_date(air_quality, 'Data Time S')
```

```python
def split_date(air_quality,col):
    time = []
    date = []
    year = []
    month = []
    day = []
    i = 0
    
    for i in range(air_quality.shape[0]):
        t = air_quality[col][i].time()
        d = air_quality[col][i].date()
        y = air_quality[col][i].date().year
        m = air_quality[col][i].date().month
        da = air_quality[col][i].date().day
        time.append(t)
        date.append(d)
        year.append(y)
        month.append(m)
        day.append(da)
        i = i+1
    
    air_quality['date'] = date
    air_quality['time'] = time
    air_quality['year'] = year
    air_quality['month'] = month
    air_quality['day'] = day
    
split_date(air_quality, 'date')
```

```python
#Kiểm tra dữ liệu"-"
columns = ['AQI index','CO', 'NO2','PM2.5', 'PM10', 'SO2', 'O3']

for col in columns:
    count = air_quality[col].value_counts()['-']
    print(f"{col}: {count}")
```

```python
#Kiểm tra dữ liệu rỗng
columns = ['AQI index','CO', 'NO2','PM2.5', 'PM10', 'SO2', 'O3']

for col in columns:
    count = air_quality[col].isna().sum()
    print(f"{col}: {count}")
```

```python
#Xử lý dữ liệu "-"
columns = ['AQI index','CO', 'NO2','PM2.5', 'PM10', 'SO2', 'O3']

for col in columns:
    air_quality[col] = air_quality[col].replace('[-]', 0, regex=True).astype(float)
```

```python
air_quality.info()
```

```python
#Xử lý dữ liệu trống
columns = ['AQI index', 'CO', 'NO2', 'PM2.5', 'PM10', 'SO2', 'O3']

for col in columns:
    # Fill missing values with 0
    air_quality[col] = air_quality[col].fillna(0)
    # Convert to float
    air_quality[col] = air_quality[col].astype(float)
```

```python
#Kiểm tra dữ liệu lại một lần nữa
columns = ['AQI index', 'CO', 'NO2', 'PM2.5', 'PM10', 'SO2', 'O3']
has_empty = False

for col in columns:
    count = air_quality[col].isnull().sum()
    if count > 0:
        has_empty = True
        print(f"{col}: {count}")

if not has_empty:
    print("Không có giá trị rỗng trong các cột được kiểm tra.")
```

```python
columns = ['AQI index', 'CO', 'NO2', 'PM2.5', 'PM10', 'SO2', 'O3']
has_dash = False

for col in columns:
    count = air_quality[col].value_counts().get('-', 0)
    if count > 0:
        has_dash = True
        print(f"{col}: {count}")

if not has_dash:
    print("Không có giá trị '-' trong các cột được kiểm tra.")
```
